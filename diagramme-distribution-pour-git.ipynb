# Diagramme de distribution d'un couple acido-basique

```python
# Import des bibliothèques 
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from matplotlib.ticker import MultipleLocator
```

```python
def diagramme(pKa):
    pH = np.arange(0,14,0.2)
    X = 10**(pH-pKa)     # X = Cb/Ca =Pb/Pa 
    Pb = X*100/(1+X)     # % de base faible 
    Pa = 100-Pb          # % d'acide faible
    
    plt.figure (figsize = (10,8))
    plt.plot(pH,Pa,"ro-",label="% acide")
    plt.plot(pH,Pb,"b+-",label="% base")
    plt.ylim(0,100)
    plt.xlim(0,14.5)
    plt.gca().xaxis.set_major_locator(MultipleLocator(1))
    plt.gca().yaxis.set_major_locator(MultipleLocator(10))
    plt.gca().xaxis.set_minor_locator(MultipleLocator(1))
    plt.gca().yaxis.set_minor_locator(MultipleLocator(10))
    plt.grid( )
    plt.xlabel("pH")
    plt.ylabel("Fraction des espèces en %")
    plt.legend()
    plt.grid()
    plt.title ("Diagramme de distribution d'un couple acide faible/base faible")
    plt.show
   
```

```python
pKa = 4.8
diagramme (pKa)
```
